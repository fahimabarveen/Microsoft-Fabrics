```sql
%%pyspark
df = spark.read.format("csv").option("header","true").load("Files/Data/Supermart Grocery Sales - Retail Analytics Dataset.csv")
# df now is a Spark DataFrame containing CSV data from "Files/Data/Supermart Grocery Sales - Retail Analytics Dataset.csv".
display(df)
```

```sql
%%pyspark
df.createOrReplaceTempView("Orders")
```

**1. Write a query to find the total sales in each region.**

```sql
%%sql
SELECT 
    Region, 
    SUM(Sales) AS Total_sales
FROM 
    Orders
GROUP BY 
    Region
ORDER BY 
Total_Sales DESC;
```

**2.	Count how many orders were placed by each customer**

```sql
%%sql
SELECT 
    Customer_Name, 
    COUNT(Customer_Name) AS Total_orders
FROM 
    Orders
GROUP BY 
   Customer_Name
ORDER BY 
    Total_orders DESC;
```

**3. Find all orders with a discount of more than 20% but less than 30%.**

```sql
%%sql
SELECT 
Order_ID,Round(SUM(Discount * 100), 2) AS Discount FROM Orders WHERE Discount > 0.2 AND Discount < 0.3 GROUP BY Order_ID;

```

**4.List the top 3 orders by profit.**

```sql
%%sql
SELECT 
    Order_ID, 
    Profit
FROM 
    Orders
ORDER BY 
    Profit DESC
LIMIT 3;
```

**5. Replace the category 'Snacks' with 'Quick Bites' in the result**

```sql
%%sql
SELECT Order_ID,
CASE
WHEN Category = 'Snacks' THEN 'Quick Bites'
ELSE Category
END AS Category
FROM
Orders;
```
